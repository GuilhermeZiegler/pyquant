```python
def sort_patterns(patterns_info):
    # Ordenar por tipo e depois por confiança
    sorted_patterns = {}
    
    # Mapeia os níveis de confiança para sua ordem
    confidence_order = {'strong': 1, 'moderate': 2, 'weak': 3}
    
    # Divide os padrões em bearish e bullish
    bearish_patterns = {k: v for k, v in patterns_info.items() if v['Type'] == 'bearish'}
    bullish_patterns = {k: v for k, v in patterns_info.items() if v['Type'] == 'bullish'}
    
    # Ordena os padrões de acordo com a confiança
    sorted_bearish = sorted(bearish_patterns.items(), key=lambda x: confidence_order[x[1]['Confidence']])
    sorted_bullish = sorted(bullish_patterns.items(), key=lambda x: confidence_order[x[1]['Confidence']])
    
    # Adiciona os padrões organizados ao dicionário final
    sorted_patterns.update(dict(sorted_bearish))
    sorted_patterns.update(dict(sorted_bullish))
    
    return sorted_patterns

# Ordena os padrões de velas
sorted_candle_patterns_info = sort_patterns(candle_patterns_info)


```
